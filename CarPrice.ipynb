## FUNCIÓN PARA SELECCIONAR COLUMNAS CON UNA CORRELACIÓN ESPECIFÍCA

Muestra los datos que se encuentran mayores a, y no importa si se repiten y con **set()** y **add** muestra los datos que se encontraron sin repetirse.

```python
def correlation(dataframe, threshold):
  cols_corr = [] #set ()
  corr_matrix = dataframe.corr()
  for i in range(len(corr_matrix.columns)):
    for j in range(i):
      if abs(corr_matrix.iloc[i, j]) >= threshold:
        col_name = corr_matrix.columns[i]
        cols_corr.append(col_name) #.add
  return cols_corr
```

```python
corr_cols = correlation(dataframe.iloc[:,:-1],0.70)
corr_cols
```

## SEPARAR COLUMNAS CATEGÓRICAS (númericas)

```python
def categorical(dataframe):
  categorical_col=[]
  for col in dataframe:
    if dataframe[col].dtype == 'object':
      categorical_col.append(col)
  return categorical_col
```

```python
categorical_col = categorical(dataframe)
print(categorical_col)
```

## OBTENER COLUMNAS NUMÉRICAS

```python
def numerical(dataframe):
  numerical_col = []
  for col in dataframe:
    if dataframe[col].dtype != 'object':
      numerical_col.append(col)
  return numerical_col
```

```python
numerical_col = numerical(dataframe)
print(numerical_col)
```
